# importing libaries


```python
import numpy as  np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.metrics import *
from  sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import AdaBoostClassifier , RandomForestClassifier
from sklearn.model_selection import train_test_split
```

Nan

```python
df = pd.read_csv('/content/census-income (1).csv', na_values = ["?", " "], skipinitialspace= True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.head()
```

```python
# labelencoding

encoder = LabelEncoder()

for x in df.columns:
  if df[x].dtypes == "object":
    df[x] = encoder.fit_transform(df[x])
```

```python
df.head()
```

```python
# model buidling

df.columns
```

```python
x = df.drop('annual_income',axis =1)
y = df['annual_income']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y , test_size =0.25, random_state = 30)
```

```python
model = AdaBoostClassifier(n_estimators = 100)
```

```python
model.fit(x_train , y_train)
```

```python
pred = model.predict(x_test)
```

```python
print(f"accuracy : {accuracy_score(y_test , pred)}")
```

```python
rfe_model = RandomForestClassifier(n_estimators = 100, random_state = 30)
rfe_model.fit(x_train , y_train)
rfe_pred = rfe_model.predict(x_test)
print(f"accuracy : {accuracy_score(y_test , rfe_pred)}")
```
